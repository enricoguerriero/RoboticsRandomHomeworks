```python
width, height = 400, 300
image_A = np.zeros((height, width, 3), dtype=np.uint8)

center_coordinates = (140, 90)  
radius = 40  
color = (255, 255, 255)  
thickness = -1  
cv2.circle(image_A, center_coordinates, radius, color, thickness)

top_left_corner = (250, 50)
bottom_right_corner = (top_left_corner[0] + 100, top_left_corner[1] + 50)
cv2.rectangle(image_A, top_left_corner, bottom_right_corner, (255, 255, 255), -1)  # -1 to fill the rectangle

cv_plot(image_A, 'Modified Image with Circle and Rectangle')
mpl_plot(image_A, 'Modified Image with Circle and Rectangle')
```

```python
width, height = 400, 300
image_A = np.zeros((height, width, 3), dtype=np.uint8)

center_coordinates = (140, 90)  
radius = 40  
color = (255, 255, 255)  
thickness = -1  
cv2.circle(image_A, center_coordinates, radius, color, thickness)

top_left_corner = (250, 50)
bottom_right_corner = (top_left_corner[0] + 100, top_left_corner[1] + 50)
cv2.rectangle(image_A, top_left_corner, bottom_right_corner, (255, 255, 255), -1)  

start_point = (100, 200)
end_point = (200, 250)
line_color = (255, 255, 255)  
line_thickness = 2
cv2.line(image_A, start_point, end_point, line_color, line_thickness)

cv_plot(image_A, 'Modified Image with Circle, Rectangle, and Line')
mpl_plot(image_A, 'Modified Image with Circle, Rectangle, and Line')
```

```python
width, height = 400, 300
image_A = np.zeros((height, width, 3), dtype=np.uint8)

center_coordinates = (140, 90)  
radius = 40  
color = (255, 255, 255)  
thickness = -1  
cv2.circle(image_A, center_coordinates, radius, color, thickness)

top_left_corner = (250, 50)
bottom_right_corner = (top_left_corner[0] + 100, top_left_corner[1] + 50)
cv2.rectangle(image_A, top_left_corner, bottom_right_corner, (255, 255, 255), -1)  

start_point = (100, 200)
end_point = (200, 250)
line_color = (255, 255, 255)  
line_thickness = 2
cv2.line(image_A, start_point, end_point, line_color, line_thickness)

triangle_points = np.array([[250, 250], [350, 250], [300, 163]], np.int32)
triangle_points = triangle_points.reshape((-1, 1, 2))
cv2.fillPoly(image_A, [triangle_points], (255, 255, 255))

cv_plot(image_A, 'Modified Image with Circle, Rectangle, Line, and Triangle')
mpl_plot(image_A, 'Modified Image with Circle, Rectangle, Line, and Triangle')
```
